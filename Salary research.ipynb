```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/lucasmazza/Downloads/Ask A Manager Salary Survey 2021 (Responses) - Form Responses 1.csv")
```

```python
#cleaning
df.columns = ["Timestamp", "Age", 
                   "Industry", "Job Title", "Context", "Salary", 
                  "Compensation","Currency", "Other Currency", 
                   "Context income",
                  "Country", "US State", "US City", 
                   "Experience overall", 
                   "Experience in Field",
                  "Edu level", "Gender", "Race"]
df
```

```python
#Which industry pays the most?
#df['Salary'].astype(str)

df['Salary'] = df['Salary'].apply(str).str.replace(',', '').astype(float)
print(df['Salary'])

salaryGrouping = df.groupby('Industry')['Salary'].mean()
salaryGrouping = salaryGrouping.to_frame().reset_index()
salaryGrouping.Salary = salaryGrouping.Salary.map(lambda x: '{:.0f}'.format(x))
salaryGrouping
```
